```python
import numpy as np
import matplotlib.pyplot as plt

c0_true = 2.5
c1_true = 1.3

n = 100
x = np.random.uniform(-5, 5, n)
y_true = c0_true + c1_true * x
noise = np.random.normal(0, 0.5, n)
y = y_true + noise

X = np.column_stack([np.ones(n), x])

A = X.T @ X
b = X.T @ y

c = np.linalg.solve(A, b)
c0_est, c1_est = c

print("True coefficients: c0 =", c0_true, ", c1 =", c1_true)
print("Estimated coefficients: c0 =", c0_est, ", c1 =", c1_est)
print("\nSystem of Linear Equations:")
print("A =\n", A)
print("b =", b)
```

```python
x_plot = np.linspace(x.min(), x.max(), 300)
y_plot = c0_est + c1_est * x_plot

plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.6, label='Data points (with noise)', s=30)
plt.plot(x_plot, y_plot, 'r-', linewidth=2, label=f'Fitted line: y = {c0_est:.3f} + {c1_est:.3f}x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression: Data and Fitted Line')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
