```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# def f(x):
#     return x**2 + 3 / 1 + x**3

def f(x):
    return 1 / (1 + 25 * x**2)

a, b = (-1, 1)
n = 15
x_points = np.linspace(a, b, n+1)
y_points = f(x_points)
print("x points:", x_points)
print("y points:", y_points)
```

```python
V = np.zeros((n + 1, n + 1))
for i in range(n + 1):
    for j in range(n + 1):
        V[i][j] = x_points[i] ** j
c = np.linalg.solve(V, y_points)
# V, c
```

```python
x_plot = np.linspace(a-0.001, b+0.001, 100)
y_func = f(x_plot)
y_plot = sum(c[i] * x_plot**i for i in range(len(c)))

fig, ax = plt.subplots()
ax.plot(x_plot, y_func, color='blue', label='Function f(x)')
ax.plot(x_plot, y_plot, color='green', label='Polynomial')
ax.scatter(x_points, y_points, color='black', label='Points')
ax.legend()
```
