```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = 1

# Spiral
# f = lambda t: r * t * np.cos(t)
# g = lambda t: r * t * np.sin(t)

# Heart shape
f = lambda t: 16 * np.power(np.sin(t), 3)
g = lambda t: 13 * np.cos(t) - 5 * np.cos(2 * t) - 2 * np.cos(3 * t) - np.cos(4 * t)

t = np.linspace(0, 10 * np.pi, 400)
x = f(t)
y = g(t)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x, y, label='Parametric plot', color='red', linewidth=2)
plt.gca().set_aspect('equal', adjustable='box')
# plt.title('Parametric Circle with r = {}'.format(r))
plt.grid()
plt.show()
```
