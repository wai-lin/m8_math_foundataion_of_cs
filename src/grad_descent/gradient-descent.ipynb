```python
def f(x):
    return (x-3)**2 + (2*x) + 3

def f_prime(x):
    return 2*(x-3)+2
```

```python
prev_x = 1
alpha = 0.1
eps = 0.0001
max_iters = 100
```

```python
x = prev_x
for i in range(max_iters):
    prev_x = x
    x -= alpha * f_prime(x)
    print(f"{i}: x = {x:.2f}, f(x) = {f(x):.2f}")

    if abs(x - prev_x) < eps:
        break

    print(f"Solution = {x:.2f}, f(x) = {f(x):.2f}, iters = {i+1}")
```
